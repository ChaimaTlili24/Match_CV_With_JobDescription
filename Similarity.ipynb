```python
from PIL import Image
import pytesseract

# Charger l'image du CV (celle que tu as uploadée par ex.)
image_path = r"C:\Users\ROYAUME MEDIAS\Downloads\archive\Resumes Datasets\Scrapped_Resumes\JavaDeveloper\bf9ade5bee889ea4.png"
img = Image.open(image_path)

# OCR pour extraire le texte en anglais
cv_text = pytesseract.image_to_string(img, lang='eng')
print(cv_text)
```

```python
blocks = [b.strip() for b in cv_text.split("\n\n") if len(b.strip()) > 5]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Bloc de référence pour détecter la section Skills
reference = ["skills technical abilities programming technologies"]

docs = blocks + reference
vectorizer = TfidfVectorizer().fit(docs)
vectors = vectorizer.transform(docs)

# Similarité entre chaque bloc et la référence
similarities = cosine_similarity(vectors[:-1], vectors[-1])
best_index = np.argmax(similarities)

skills_section = blocks[best_index]
print("=== Bloc Skills détecté ===")
print(skills_section)
```

```python
# Nettoyer le texte OCR
cv_text_clean = cv_text.replace('\x0c', '\n')  # supprimer le caractère de fin de page
cv_text_clean = cv_text_clean.replace('\r', '\n')
cv_text_clean = "\n".join([l.strip() for l in cv_text_clean.split("\n") if l.strip() != ""])

# Chercher la ligne contenant 'skill' et prendre toutes les lignes suivantes jusqu'à une section probable
lines = cv_text_clean.split("\n")
skills_list = []
capture = False

for line in lines:
    # Si la ligne contient 'skill', commencer la capture
    if "skill" in line.lower():
        capture = True
        continue
    if capture:
        # Arrêter si ligne vide ou nouvelle section (majuscules, ":" ou "Experience", "Education", etc.)
        if line == "" or line.endswith(":") or any(word in line.lower() for word in ["experience", "education", "certificates", "projects"]):
            break
        skills_list.append(line)

# Filtrer pour enlever les langues et le mot "Languages"
words_to_ignore = ["languages", "english", "japanese", "french", "spanish", "german", "arabic"]
skills_filtered = [s for s in skills_list if s.lower() not in words_to_ignore]

print("\n=== Compétences détectées ===")
print(skills_filtered)
```
